```python
def reassign_miscellaneous(row):
    type_val = str(row['Type'])
    sub_category = str(row['Categories: Sub'])
    product_group = str(row['Product Group'])
    
    if product_group == 'Grocery':
        return 'Grocery'
    elif product_group in ['Home', 'Home Improvement', 'Kitchen', 'Furniture']:
        return 'Home & Kitchen'
    elif product_group == 'Pet Products':
        return 'Pet Supplies'
    elif 'Sport' in type_val or 'Outdoor' in type_val:
        return 'Sports & Outdoors'
    elif product_group in ['Apparel', 'Shoes', 'Jewelry']:
        return 'Clothing, Shoes & Jewelry'
    elif product_group in ['Beauty', 'Drugstore']:
        return 'Beauty & Personal Care'
    elif product_group in ['Toy', 'Video Games', 'Music', 'Office Product']:
        return 'Hobby & Leisure'
    elif 'Baby' in type_val:
        return 'Baby Products'
    elif product_group in ['Electronics', 'Major Appliances', 'Wireless']:
        return 'Electronics & Home Appliances'
    else:
        return 'Miscellaneous'

# Apply the reassignment logic to the Miscellaneous products
data.loc[data['New_Category'] == 'Miscellaneous', 'New_Category'] = data[data['New_Category'] == 'Miscellaneous'].apply(reassign_miscellaneous, axis=1)

# Display the distribution of the new categories after reassignment
new_category_counts_after_reassignment = data['New_Category'].value_counts()
new_category_counts_after_reassignment
```
