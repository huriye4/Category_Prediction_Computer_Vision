```python

def indir(url, dosya_adi):
    """
    Belirtilen URL'den bir resmi indirir ve belirtilen dosya adıyla kaydeder.

    Args:
        url (str): İndirilecek resmin URL'si.
        dosya_adi (str): Resmin kaydedileceği dosya adı.
    """
    resim_cevabi = requests.get(url)
    if resim_cevabi.status_code == 200:
        with open(dosya_adi, 'wb') as f:
            f.write(resim_cevabi.content)
    else:
        print(f"Hata: {url} adresinden resim indirilemedi ({resim_cevabi.status_code})")

# URL'lerinizi içeren bir liste oluşturun
url_ler =df["URL: Keepa"]

# Her URL için bir resim indirin
for i, url in enumerate(url_ler):
    dosya_adi = f"resim_{i+1}.jpg"
    indir(url, dosya_adi)
    print(f"{i+1}/{len(url_ler)} resim indirildi.")
```

```python
import os
import requests
import pandas as pd
import time

def indir(url, dosya_adi):
    """
    Belirtilen URL'den bir resmi indirir ve belirtilen dosya adıyla kaydeder.

    Args:
        url (str): İndirilecek resmin URL'si.
        dosya_adi (str): Resmin kaydedileceği dosya adı.
    """
    try:
        resim_cevabi = requests.get(url, timeout=10)  # Timeout süresi 10 saniye olarak ayarlandı
        resim_cevabi.raise_for_status()  # HTTP hatalarını kontrol eder
        with open(dosya_adi, 'wb') as f:
            f.write(resim_cevabi.content)
        return True
    except requests.exceptions.RequestException as e:
        print(f"Hata: {url} adresinden resim indirilemedi. Hata: {e}")
        return False

# Veri setini yükle
df = pd.read_excel("C:\\Users\\w10\\OneDrive\\Masaüstü\\Internship GitHub\\i-brands\\ML\\Scraping\\Data\\Walmart\\walmart_item_report.xlsx")

# URL'lerinizi içeren bir liste oluşturun
url_ler = df["Primary Image URL"]

# Her URL için bir resim indirin
for i, url in enumerate(url_ler):
    
    # Geçersiz dosya adı karakterlerini temizle
    kategori = "".join(x for x in df["Product Category"][i] if x.isalnum() or x in "._-")
    dosya_adi = f"{kategori}_{i}.jpg"
    
    for _ in range(3):  # 3 defa yeniden dene
        if indir(url, dosya_adi):
            print(f"{i+1}/{len(url_ler)} resim indirildi.")
            break  # İndirme başarılıysa döngüden çık
        else:
            print(f"Hata: {i+1}. resim indirilemedi. 5 saniye bekleyerek tekrar denenecek.")
            time.sleep(5)  # 5 saniye bekleyerek tekrar dene
```
